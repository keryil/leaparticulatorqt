```python
import pymc as pm
from pymc.Matplot import plot
discrete_sum = sum(sum(d) for d in discrete_counts)
continuous_sum = sum(sum(d) for d in continuous_counts)
discrete_avg = float(discrete_sum) / sum(len(l) for l in discrete_counts)
discrete_n = sum(len(l) for l in discrete_counts)
continuous_avg = float(continuous_sum) / sum(len(l) for l in continuous_counts)
continuous_n = sum(len(l) for l in continuous_counts)
print discrete_sum, discrete_avg
print continuous_sum, continuous_avg
```

```python
p_iconicity_continuous = pm.Uniform("Cont. iconicity prob",0,1)
iconicity_continuous = pm.Binomial("Iconicity binomial cont", p=p_iconicity_continuous, n=continuous_n,
                                   observed=True, value=continuous_sum)

p_iconicity_discrete = pm.Uniform("Disc. iconicity prob",0,1)
iconicity_disc = pm.Binomial("Iconicity binomial disc", p=p_iconicity_discrete, n=discrete_n,
                                   observed=True, value=discrete_sum)
@pm.deterministic(plot=True,trace=True)
def difference(cont=p_iconicity_continuous, disc=p_iconicity_discrete):
    return cont-disc
```

```python
m = pm.Model([difference, iconicity_disc, iconicity_continuous, 
              p_iconicity_continuous, p_iconicity_discrete])
posteriori = pm.MAP(m)
posteriori.fit()
mcmc = pm.MCMC(m)


mcmc.sample(400000, 200000,1)
plot(mcmc)
```

```python
discrete_sums = [int(sum(counts)) for counts in discrete_counts]
cont_sums = [int(sum(counts)) for counts in continuous_counts]
print discrete_sums, cont_sums

p_continuous = []
p_discrete = []
iconicity_continuous = []
iconicity_discrete = []
differences_continuous = []
differences_discrete = []
all_vars = []
for phase in range(3):
    p_cont = pm.Uniform("Cont. iconicity p, phase%d" % phase, 0, 1, plot=False)
    p_continuous.append(p_cont)
    all_vars.append(p_cont)
    p_disc = pm.Uniform("Disc. iconicity p, phase%d" % phase, 0, 1, plot=False)
    p_discrete.append(p_disc)
    all_vars.append(p_disc)
    i_cont = pm.Binomial("Iconicity binomial cont phase%d" % phase, p=p_continuous[-1], n=len(continuous_counts[phase]),
                                   observed=True, value=cont_sums[phase])
    iconicity_continuous.append(i_cont)
    all_vars.append(i_cont)
    i_disc = pm.Binomial("Iconicity binomial disc phase%d" % phase, p=p_discrete[-1], n=len(discrete_counts[phase]),
                                   observed=True, value=discrete_sums[phase])
    iconicity_discrete.append(i_disc)
    all_vars.append(i_disc)
    for previous_phase in range(phase):
        @pm.deterministic(plot=True,trace=True,name="Disc phase%d m phase%d" % (phase, previous_phase))
        def disc_diff(this_phase=iconicity_discrete[phase], last_phase=iconicity_discrete[previous_phase]):
            return this_phase - last_phase
        @pm.deterministic(plot=True,trace=True,name="Cont phase%d m phase%d" % (phase, previous_phase))
        def cont_diff(this_phase=iconicity_continuous[phase], last_phase=iconicity_continuous[previous_phase]):
            return this_phase - last_phase
        
        differences_discrete.append(disc_diff)
        differences_continuous.append(cont_diff)
        
        all_vars.extend([disc_diff, cont_diff])
        
m = pm.Model(all_vars)
posterior = pm.MAP(m)
posterior.fit
mcmc = pm.MCMC(m)
mcmc.sample(400000, 200000,1)
```
