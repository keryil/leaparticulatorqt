```python
import numpy as np
from __future__ import division
```

# The implementation

```python
def population_to_relative_freq(population):
    return population / sum(population)

def step(ecm, population):
    relative_freqs = np.asarray([0. for i in range(len(ecm))])
    # the denominator, shared accross all types
    denom = 0.
    
    # convert population counts to relative frequencies
    pop = population_to_relative_freq(population)
    
    # for each type
    for i, freq in enumerate(pop):
        # for each type, again
        for a in range(len(pop)):
#             print "a: %d, i: %d" % (a,i)
            new = freq * ecm[i,a]
#             print "freq: %f * ecm[i,a]: %f = %f" % (freq, ecm[i,a], new)
            denom += new
    
#     print "Denom", denom
    
    # for each type
    for i in range(len(pop)):
        numerator = 0.
        for j, freq in enumerate(pop):
#             print "freq: %f * ecm[j,i]: %f = %f" % (freq, ecm[j,i], new)
            numerator += freq * ecm[j,i]
#         print "Num", numerator
        relative_freqs[i] = float(numerator) / float(denom)
#         print "New freq", relative_freqs[i]
        
    return population_to_relative_freq(relative_freqs)
        
```

```python
def simulate(initial_pop, ecm, generations):
    population_trace = [population_to_relative_freq(initial_pop)]
    population = initial_pop
    for i in range(generations):
        population = step(ecm, population)
        population_trace.append(population)
    return population_trace
    
# population = np.asarray([999,1])
# print population_to_relative_freq(population)
# trace = simulate(population, ecm, 40)
```

# Dar-ta vs. Day-ta (evolutionary, equal homoimpacts, type benefitting from greater heteroimpact dominates)

```python
darta = np.asarray([5,4])
dayta = np.asarray([.1,5])
ecm = np.asmatrix(np.asarray([darta,dayta]))
population = np.asarray([1-0.001,.001])
print "ECM:\n", ecm
print "Population:\n", population_to_relative_freq(population)
trace = simulate(population, ecm, 40)
```

```python
darta_t, dayta_t = zip(*trace)
from matplotlib import pyplot as plt
plt.plot(darta_t, label="dar-ta")
plt.plot(dayta_t, label="day-ta")
plt.legend()
```

# Reproduction with rare mutations (almost a replicator, type with higher reproduction rate dominates)

```python
type_a = np.asarray([4,.001])
type_b = np.asarray([.001,5])
ecm = np.asmatrix(np.asarray([type_a,type_b]))
population = np.asarray([1-.0005,.0005])
print "ECM:\n", ecm
print "Population:\n", population_to_relative_freq(population)
trace = simulate(population, ecm, 40)
```

```python
type_a, type_b = zip(*trace)
from matplotlib import pyplot as plt
plt.plot(type_a, label="type A")
plt.plot(type_b, label="type B")
plt.legend()
```

# The Folktale (no homoimpact, oscillates based on initial frequencies)

```python
mr = np.asarray([0,6])
pt = np.asarray([.2,0])
ecm = np.asmatrix(np.asarray([mr,pt]))
small_one = 0.1
population = np.asarray([1-small_one,small_one])
print "ECM:\n", ecm
print "Population:\n", population_to_relative_freq(population)
trace = simulate(population, ecm, 30)
```

```python
mr, pt = zip(*trace)
from matplotlib import pyplot as plt
plt.plot(mr, label="MR")
plt.plot(pt, label="PT")
plt.legend()
```
